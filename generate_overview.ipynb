```python
import numpy as np
import pandas as pd
```

```python
filename = 'mock_data.csv'
df = pd.read_csv(filename, sep=';')
df
```

```python
df_2d = df.merge(df, on=['test_id', 'status'], how='outer')\
    .sort_values(['category_x', 'value_x', 'category_y', 'value_y'])\
    .groupby(['category_x', 'value_x', 'category_y', 'value_y'])\
    .count()\
    .unstack(level=[3,2])\
    .replace(np.nan, 0)
```
